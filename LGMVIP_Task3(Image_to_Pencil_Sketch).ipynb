**IMPORTING THE REQUIRED LIBRARIES**

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

**READING THE IMAGE IN BGR FORMAT**

```python
image = cv2.imread("/content/Khaomanee_cat.jpg")
o_image =cv2.resize(image, (500, 800))
cv2_imshow(o_image)
```

**CONVERTING A BGR IMAGE INTO RGB IMAGE**

```python
rgbimg = cv2.cvtColor(o_image, cv2.COLOR_BGR2RGB)
cv2_imshow(rgbimg)
```

**CONVERTING RGB IMAGE INTO GRAYSCALE IMAGE**

```python
grayimg = cv2.cvtColor(rgbimg, cv2.COLOR_RGB2GRAY)
cv2_imshow(grayimg)
```

**INVERTING THE IMAGE**

```python
invertimg = cv2.bitwise_not(grayimg)
cv2_imshow(invertimg)
```

**BLURRING THE IMAGE**

```python
blurimg= cv2.GaussianBlur(invertimg, (15, 15), sigmaX=0, sigmaY=0)
cv2_imshow(blurimg)
```

**PENCIL SKETCH**

```python
pencil_sketch = cv2.divide(grayimg, 255 - blurimg, scale=256)
cv2_imshow(pencil_sketch)
```

**ORIGNAL IMAGE AND PENCIL SKETCH**

```python
cv2_imshow(o_image)
cv2_imshow(pencil_sketch)
```

The cat image is converted into pencil sketch with cv2 library in python.
